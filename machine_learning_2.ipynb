```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
img = Image.open('./for_conv.png').convert('RGB')
fig = plt.figure(figsize=(15, 15))
plt.subplot(3, 1, 1)
plt.imshow(img)

input_tensor = torchvision.transforms.ToTensor()(img)
input_tensor = input_tensor.unsqueeze(0)

conv = torch.nn.Conv2d(3, 1, 3)
relu = torch.nn.ReLU()

conv_output = conv(input_tensor)
conv_img = conv_output.squeeze(0)
conv_img = torchvision.transforms.ToPILImage()(conv_img)

plt.subplot(3, 1, 2)
cmap1 = plt.imshow(conv_img)
fig.colorbar(cmap1)

relu_output = relu(conv_output)
relu_img = relu_output.squeeze(0)
relu_img = torchvision.transforms.ToPILImage()(relu_img)


plt.subplot(3, 1, 3)
cmap2 = plt.imshow(relu_img)
fig.colorbar(cmap2)
```
