```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
img = Image.open('/home/scidev/screenshots/for_conv.png').convert('RGB')
plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
plt.imshow(img)

img = torchvision.transforms.ToTensor()(img)

img = img.unsqueeze(0)
conv = torch.nn.Conv2d(3, 3, 3)

img = conv(img)

img = img.squeeze(0)

type(img)

img = torchvision.transforms.ToPILImage()(img)

plt.subplot(2, 1, 2)
plt.imshow(img)
```
